```python
#1
from sklearn.datasets import load_iris
from sklearn.model_selection import cross_validate
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import make_scorer, f1_score
import time

# Wczytaj zbiór danych IRIS
iris = load_iris()
X, y = iris.data, iris.target
knn = KNeighborsClassifier(n_neighbors=3)

# A. Metoda cross_validate dla 5 podzbiorów
cv_results = cross_validate(knn, X, y, cv=5, n_jobs=4, verbose=1, return_train_score=True)

# B. Wypisz wynik testu dla każdego z podzbiorów
print("Wyniki testu dla każdego podzbioru:", cv_results['test_score'])

# C. Użyj metryki F1 score
f1 = make_scorer(f1_score, average='macro')
cv_results = cross_validate(knn, X, y, cv=5, n_jobs=4, verbose=1, scoring=f1, return_train_score=True)
print("Średni wynik F1:", cv_results['test_score'].mean())

# D. Wypisz wynik testu F1 score dla każdego z podzbiorów
print("Wyniki F1 dla każdego podzbioru:", cv_results['test_score'])

# E. Wypisz wyniki testu dla metryki F1 score dla każdego z podzbiorów i jego przeciętną wartość
print("Wyniki F1 dla każdego podzbioru:", cv_results['test_score'])
print("Średni wynik F1:", cv_results['test_score'].mean())

# F. Użyj walidacji krzyżowej z 10 podzbiorami, czy wynik zmieni się?
cv_results_10 = cross_validate(knn, X, y, cv=10, n_jobs=4, verbose=1, scoring=f1, return_train_score=True)
print("Wyniki F1 dla każdego podzbioru (10 podzbiorów):", cv_results_10['test_score'])
print("Średni wynik F1 (10 podzbiorów):", cv_results_10['test_score'].mean())

# G. Zmierzy czas treningu modelu
start_time = time.time()
cross_validate(knn, X, y, cv=5, n_jobs=4, verbose=1, scoring=f1, return_train_score=True)
end_time = time.time()
print("Czas treningu (5 podzbiorów):", end_time - start_time)

start_time = time.time()
cross_validate(knn, X, y, cv=10, n_jobs=4, verbose=1, scoring=f1, return_train_score=True)
end_time = time.time()
print("TCzas treningu (10 podzbiorów", end_time - start_time)
```

```python
#2
from sklearn.datasets import load_iris
from sklearn.model_selection import GridSearchCV, cross_validate
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import make_scorer, f1_score
import pandas as pd
import time

# Wczytaj zbiór danych IRIS
iris = load_iris()
X, y = iris.data, iris.target

# A. Przygotuj model z użyciem GridSearch
param_grid = {
    'max_depth': [1, 2, 4, 8, 16, 32, 64],
    'n_estimators': [1, 2, 3, 4, 5, 10, 20, 25]
}
rf = RandomForestClassifier(random_state=42)
grid_search = GridSearchCV(rf, param_grid, cv=5, scoring='f1_macro', n_jobs=4, verbose=1)
grid_search.fit(X, y)

# B. Wypisz najlepsze parametry
print("Najlepsze parametry:", grid_search.best_params_)

# C. Wypisz wynik testu dla każdej z kombinacji parametrów
results = pd.DataFrame(grid_search.cv_results_)
print("Wyniki testu dla każdej z kombinacji parametrów:\n", results[['params', 'mean_test_score', 'std_test_score']])

# D. Podaj przeciętny wynik dla testu z użyciem metryki F1 score
print("Średni wynik F1 dla najlepszych parametrów:", grid_search.best_score_)

# E. Wypisz wynik testu F1 score dla każdego z podzbiorów
best_model = grid_search.best_estimator_
cv_results = cross_validate(best_model, X, y, cv=5, n_jobs=4, verbose=1, scoring='f1_macro', return_train_score=True)
print("Wyniki F1 dla każdego podzbioru:", cv_results['test_score'])

# F. Wypisz wyniki testu dla metryki F1 score dla każdego z podzbiorów i jego przeciętną wartość
print("Wyniki F1 dla każdego podzbioru:", cv_results['test_score'])
print("Średni wynik F1:", cv_results['test_score'].mean())

# G. Użyj walidacji krzyżowej z 10 podzbiorami, czy wynik zmieni się?
cv_results_10 = cross_validate(best_model, X, y, cv=10, n_jobs=4, verbose=1, scoring='f1_macro', return_train_score=True)
print("Wyniki F1 dla każdego podzbioru (10 podzbiorów):", cv_results_10['test_score'])
print("Średni wynik F1 (10 podzbiorów):", cv_results_10['test_score'].mean())

# Zmierzy czas treningu modelu
start_time = time.time()
cross_validate(best_model, X, y, cv=5, n_jobs=4, verbose=1, scoring='f1_macro', return_train_score=True)
end_time = time.time()
print("Czas treningu (5 podzbiorów):", end_time - start_time)

start_time = time.time()
cross_validate(best_model, X, y, cv=10, n_jobs=4, verbose=1, scoring='f1_macro', return_train_score=True)
end_time = time.time()
print("Czas treningu (10 podzbiorów):", end_time - start_time)
```

```python
#8
import numpy as np
import pandas as pd
from sklearn.datasets import make_regression
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import RandomizedSearchCV, cross_validate
from sklearn.metrics import make_scorer, mean_squared_error
import time

# Generowanie danych
X, y = make_regression(n_samples=400, n_features=4, noise=0.1, random_state=42)

# A. Przygotuj model z użyciem RandomizedSearch
param_distributions = {
    'max_depth': [1, 2, 4, 8, 16, 32, 64, 128, 256, 512, 1024],
    'n_estimators': [3, 6, 9, 12, 15, 18, 21, 24, 27, 30]
}
rf = RandomForestRegressor(random_state=42)
random_search = RandomizedSearchCV(rf, param_distributions, n_iter=100, cv=5, scoring='neg_mean_squared_error', n_jobs=4, verbose=1, random_state=42)
random_search.fit(X, y)

# B. Wypisz najlepsze parametry
print("Najlepsze parametry:", random_search.best_params_)

# C. Wypisz wynik testu dla każdej z kombinacji parametrów
results = pd.DataFrame(random_search.cv_results_)
print("Wyniki testu dla każdej z kombinacji parametrów:\n", results[['params', 'mean_test_score', 'std_test_score']])

# D. Podaj przeciętny wynik dla testu z użyciem metryki F1 score
print("Średni wynik F1 dla najlepszych parametrów:", random_search.best_score_)

# E. Wypisz wynik testu F1 score dla każdego z podzbiorów
best_model = random_search.best_estimator_
mse = make_scorer(mean_squared_error)
cv_results = cross_validate(best_model, X, y, cv=5, n_jobs=4, verbose=1, scoring=mse, return_train_score=True)
print("Wyniki MSE dla każdego podzbioru:", cv_results['test_score'])

# F. Użyj walidacji krzyżowej z 10 podzbiorami, czy wynik zmieni się?
cv_results_10 = cross_validate(best_model, X, y, cv=10, n_jobs=4, verbose=1, scoring=mse, return_train_score=True)
print("Wyniki MSE dla każdego podzbioru (10 podzbiorów):", cv_results_10['test_score'])
print("Średni wynik MSE (10 podzbiorów):", cv_results_10['test_score'].mean())

# Zmierz czas treningu modelu
start_time = time.time()
cross_validate(best_model, X, y, cv=5, n_jobs=4, verbose=1, scoring=mse, return_train_score=True)
end_time = time.time()
print("Czas treningu (5 podzbiorów):", end_time - start_time)

start_time = time.time()
cross_validate(best_model, X, y, cv=10, n_jobs=4, verbose=1, scoring=mse, return_train_score=True)
end_time = time.time()
print("Czas treningu (10 podzbiorów):", end_time - start_time)
```

```python
#10
import numpy as np
import pandas as pd
from sklearn.datasets import make_regression
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import RandomizedSearchCV, cross_validate
from sklearn.metrics import make_scorer, r2_score

# Generowanie danych
X, y = make_regression(n_samples=400, n_features=4, noise=0.1, random_state=42)

# Model
rf = RandomForestRegressor(random_state=42)

# Definicja siatki parametrów do przeszukania
param_distributions = {
    'max_depth': [1, 2, 4, 8, 16, 32, 64, 128, 256, 512, 1024],
    'n_estimators': [3, 6, 9, 12, 15, 18, 21, 24, 27, 30]
}

# Przeszukiwanie losowe
random_search = RandomizedSearchCV(rf, param_distributions, n_iter=100, cv=5, scoring='r2', n_jobs=4, verbose=1, random_state=42)
random_search.fit(X, y)

# Wyświetlenie najlepszych parametrów i wyników
print("Best parameters:", random_search.best_params_)
results = pd.DataFrame(random_search.cv_results_)
print(results[['params', 'mean_test_score', 'std_test_score']])
print("Average R^2 score for best parameters:", random_search.best_score_)

best_model = random_search.best_estimator_

# Ocena modelu za pomocą walidacji krzyżowej
cv_results = cross_validate(best_model, X, y, cv=5, n_jobs=4, scoring='r2', return_train_score=True)
print("R^2 scores for each fold:", cv_results['test_score'])
print("Average R^2 score:", cv_results['test_score'].mean())

cv_results_10 = cross_validate(best_model, X, y, cv=10, n_jobs=4, scoring='r2', return_train_score=True)
print("R^2 scores for each fold (10-fold):", cv_results_10['test_score'])
print("Average R^2 score (10-fold):", cv_results_10['test_score'].mean())
```
