```python
#2
from sklearn.datasets import load_iris
from collections import Counter

dataset_iris = load_iris()

print("Nazwy kolumn:", dataset_iris.feature_names)

nazwy_gatunkow = dataset_iris.target_names
print("Nazwy badanych gatunk√≥w kwiatk√≥w:", nazwy_gatunkow)

target_counts = Counter(dataset_iris.target)
for gatunek, count in zip(nazwy_gatunkow, target_counts.values()):
    print(f"Ilo≈õƒá przypadk√≥w dla gatunku {gatunek}: {count}")
```
